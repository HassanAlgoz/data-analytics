# Feature Scaling

```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/car_prices/CarPrices.csv')

df.head()
```

```python
df.describe().T.loc['peakrpm']
```

### Max-scaling

```python
max_scaled = df['peakrpm'] / df['peakrpm'].max()
max_scaled.describe()
```

### Min-max scaling

```python
minmax_scaled = (df['peakrpm'] - df['peakrpm'].min() ) / (df['peakrpm'].max() - df['peakrpm'].min())
minmax_scaled.describe()
```

### Standardization

```python
standardized = (df['peakrpm'] - df['peakrpm'].mean() ) / df['peakrpm'].std()
```

```python
# subplots showing the original and the scaled data
import matplotlib.pyplot as plt


fig, ax = plt.subplots(4, 1, figsize=(10, 10))
fig.subplots_adjust(hspace=0.5)

df['peakrpm'].plot(ax=ax[0], title='Original Data')
max_scaled.plot(ax=ax[1], title='Max Scaled Data')
minmax_scaled.plot(ax=ax[2], title='MinMax Scaled Data')
standardized.plot(ax=ax[3], title='Standardized Data')
```
