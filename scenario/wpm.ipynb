# Words-per Minute

EDA done step-by-step on WPM scores from a typing test.

Objectives:

- Learn Univariate analysis
- Learn how categorical variable is used in the analysis of numerical variables

```python
more_scores = [46, 23, 36, 21, 38, 23, 22, 27, 28, 37, 37, 26, 33, 53, 9, 13, 44]
```

```python
scores = [
    33,
    46,
    35,
    28,
    20,
    22,
    19,
    25,
    19,
    67,
    23,
    19,
    # female
    40,
    18,
    27,
]
```

```python
s = 0
for i in scores:
    s += i
s / len(scores)
```

```python
mean = sum(scores) / len(scores)
mean # Measures Central Tendency
```

```python
sorted(scores)
```

```python
mid_index = len(scores) // 2
median = sorted(scores)[mid_index]
median # Measures Central Tendency
```

```python
import pandas as pd

print(type(scores))
s = pd.Series(scores, name="wpm")
print(type(s))
```

```python
s.std() # Measure of Dispersion (Variance)
```

```python
len(s)
```

```python
from matplotlib import pyplot as plt

s.plot.hist(bins=15)
plt.title("Words per minute (WPM)")
```

```python
from matplotlib import pyplot as plt

s.plot.box()
plt.title("Words per minute (WPM)")
```
