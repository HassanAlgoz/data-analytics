```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)
```

## Nominal Data

```python
# Example 1: Ice-Cream Flavors
flavors = ["Vanilla", "Chocolate", "Strawberry", "Blueberry"]
flavor_data = np.random.choice(flavors, size=100, p=[0.6, 0.3, 0.075, 0.025])
df = pd.DataFrame(flavor_data, columns=["Flavor"])
df
```

```python
# Count values in each category
counts = df["Flavor"].value_counts()
counts
```

A good practice in making the bar chart is to sort the bars by `counts`, since it makes reading the bar chart easier.

Note: `value_counts()` results are, by default, sorted (in descending order).

```python
# Create a bar chart of flavor counts
plt.figure(figsize=(6, 3))
plt.bar(counts.index, counts.values)
plt.xlabel("Hair color")
plt.ylabel("Number of people")
plt.title("Hair color distribution")
plt.show()
```

We can also specify colors for each bar, based on the flavor:

```python
colors = [
    "#f4e5b7",
    "#9d7a5d",
    "#b73924",
    "#000000",
    "#464196",
]
# Create a bar chart of flavor counts
plt.figure(figsize=(4, 2))
plt.bar(counts.index, counts.values, color=colors)
plt.ylabel("Count")
plt.title("Flavor distribution")
plt.show()
```

We can also draw them horizontally:

```python
# Create a barh (horizontal) chart of flavor counts
plt.figure(figsize=(4, 2))
plt.barh(counts.index, counts.values, color=colors)
plt.xlabel("Count")
plt.title("Flavor distribution")
plt.show()
```

### Pie Charts

Pie charts are not really recommended in practice, especially when the number of categories gets big:

```python
# Create a bar chart of the counts
plt.figure(figsize=(4, 5))
plt.pie(counts, colors=colors, labels=counts.index, autopct="%1.1f%%", textprops={"fontsize": 12, "weight": "bold"})
plt.title("Flavor distribution")
plt.show()
```

## Ordinal Data

```python
# Create a pandas DataFrame with odrinal data

df = pd.DataFrame({
    "Rating": np.random.randint(1, 6, 100)
})
df
```

```python
# Count values in each category
counts = df["Rating"].value_counts()
counts
```

```python
# Create a bar chart of the satisfaction rating counts
plt.figure(figsize=(4, 3))
plt.bar(counts.index, counts.values)
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Customer satisfaction rating distribution")
plt.show()
```

## Visualize Numerical Data

```python
x = np.linspace(0, 2, 100)  # data points: start=0, end=2, points=100
x
```

```python
plt.figure(figsize=(5, 2.5), layout="constrained")

y = x
plt.plot(x, y, label="linear")

y = x ** 2
plt.plot(x, y, label="quadratic")

y = x ** 3
plt.plot(x, y, label="cubic")

plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Line Chart")
plt.legend()
```
