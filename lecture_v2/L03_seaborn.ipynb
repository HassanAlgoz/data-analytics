# Seaborn Fundamentals

[Seaborn](https://seaborn.pydata.org/index.html) is a library for making statistical graphics in Python. It builds on top of [matplotlib](https://matplotlib.org/) and integrates closely with pandas data structures.

Matplotlib features:

- Create publication quality plots.
- Make interactive figures that can zoom, pan, update.
- Customize visual style and layout.
- Export to many file formats.
- Embed in JupyterLab and Graphical User Interfaces.
- Use a rich array of third-party packages built on Matplotlib.

Seaborn categorizes its plots into the following types:

- Relational plots
- Distribution plots
- Categorical plots
- Regression plots

It also has features for:

- Multi-plot grids
- Figure theming
- Color palettes

We **highly recommend** going first through the introduction: https://seaborn.pydata.org/tutorial/introduction.html to familiarize your self with `sns` API in general, and where to find stuff as you might need them later.

You may also look at the following Kaggle course on Data Visualization: https://www.kaggle.com/learn/data-visualization

## Data to Viz

There are two considerations when making plots:

1. **The What**: what plot is most suitable for the data + goal of the plot
2. **The How**: the sequence of steps to make the plot render and show up on the screen (or save to file)

With [Data-to-Viz](https://www.data-to-viz.com): You pick the main data type then let the decision tree guide you toward your graphic possibilities. Alternatively, check the complete decision tree.
